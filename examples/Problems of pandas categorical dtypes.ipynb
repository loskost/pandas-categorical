```python
import numpy as np
import pandas as pd
```

# read dataframe

```python
# optimal column types
SUB_DTYPES = {
    'Invoice/Item Number': "string",
    'Date': 'datetime64[ns]',
    'Store Number': "uint16",
    'Store Name': "string",
    'Address': 'string',
    'City': "string",
    'Zip Code': 'string',
    'Store Location': 'string',
    'County Number': "Int16",
    'County': "string",
    'Category': "Int32",
    'Category Name': 'string',
    'Vendor Number': "Int16",
    'Vendor Name': 'string',
    'Item Number': "string",
    'Item Description': "string",
    'Pack': "uint16",
    'Bottle Volume (ml)': "uint32",
    'State Bottle Cost': float,
    'State Bottle Retail': float,
    'Bottles Sold': int,
    'Sale (Dollars)': float,
    'Volume Sold (Liters)': float,
    'Volume Sold (Gallons)': float,
}
```

```python
df = pd.read_csv(
    "/kaggle/input/big-sales-data/Sales_Data/Liquor_Sales.csv",
    engine='pyarrow',
    dtype=SUB_DTYPES
)
```

```python
# Let's put a small part of the data aside
(_, df), (_, df_1) = df.groupby(df.index < 1000)
```

# Compression by categorization

```python
df.info()
```

```python
CAT_COLS = [
    'Date',
    'Store Number',
    'Store Name',
    'Address',
    'City',
    'Zip Code',
    'Store Location',
    'County Number',
    'County',
    'Category',
    'Category Name',
    'Vendor Number',
    'Vendor Name',
    'Item Number',
    'Item Description',
    'Pack',
    'Bottle Volume (ml)',
]

for cat_col in CAT_COLS:
    print(cat_col)
    df[cat_col] = df[cat_col].astype("category")
```

```python
df.info()
```

# Loss of categorical types in concat

```python
df = pd.concat([df, df_1], ignore_index=True)
```
